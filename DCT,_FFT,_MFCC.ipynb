# Zero frequency

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform DCT and shift zero-frequency component to the center
def perform_dct_with_center_shift(image_path):
    # Load the image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"Image at {image_path} could not be loaded.")

    # Resize the image to 224x224 (size expected by ViT model)
    img = cv2.resize(img, (224, 224))

    # Normalize pixel values to range [0, 1]
    img_normalized = img.astype(np.float32) / 255.0

    # Perform DCT
    dct_img = cv2.dct(img_normalized)

    # Shift the zero-frequency component to the center
    dct_img_centered = np.fft.fftshift(dct_img)

    return dct_img_centered

# Function to visualize an image with enhanced contrast
def visualize_image(image, title):
    # Apply normalization to the DCT image
    normalized_img = cv2.normalize(image, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
    normalized_img = np.clip(normalized_img, 0, 255).astype(np.uint8)

    plt.imshow(normalized_img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.colorbar()
    plt.show()

# Paths to your real and fake images
real_image_path = 'real.jpg'
fake_image_path = 'fake.jpg'

# Perform DCT with center shift on both images
real_dct_img_centered = perform_dct_with_center_shift(real_image_path)
fake_dct_img_centered = perform_dct_with_center_shift(fake_image_path)

# Visualize the centered DCT images
visualize_image(real_dct_img_centered, "Centered DCT Transformed Real Image")
visualize_image(fake_dct_img_centered, "Centered DCT Transformed Fake Image")
```

Centering High-Frequency Components

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform DCT and shift high-frequency components to the center
def perform_dct_high_freq_center(image_path):
    # Load the image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"Image at {image_path} could not be loaded.")

    # Resize the image to 224x224 (size expected by ViT model)
    img = cv2.resize(img, (224, 224))

    # Normalize pixel values to range [0, 1]
    img_normalized = img.astype(np.float32) / 255.0

    # Perform DCT
    dct_img = cv2.dct(img_normalized)

    # Shift the zero-frequency component to the center
    dct_img_centered = np.fft.fftshift(dct_img)

    # High-pass filter: Retain only high-frequency components
    mask = np.zeros(dct_img_centered.shape, dtype=np.float32)
    center = dct_img_centered.shape[0] // 2
    mask[:center - 50, :center - 50] = 1
    mask[center + 50:, :center - 50] = 1
    mask[:center - 50, center + 50:] = 1
    mask[center + 50:, center + 50:] = 1

    dct_img_high_freq_centered = dct_img_centered * mask

    return dct_img_high_freq_centered

# Function to normalize images globally
def global_normalize(images):
    all_min = min(np.min(image) for image in images)
    all_max = max(np.max(image) for image in images)

    normalized_images = []
    for image in images:
        normalized_img = (image - all_min) / (all_max - all_min) * 255
        normalized_img = np.clip(normalized_img, 0, 255).astype(np.uint8)
        normalized_images.append(normalized_img)

    return normalized_images

# Function to enhance contrast using histogram equalization
def enhance_contrast(images):
    enhanced_images = []
    for image in images:
        enhanced_img = cv2.equalizeHist(image)
        enhanced_images.append(enhanced_img)
    return enhanced_images

# Function to visualize an image
def visualize_image(image, title):
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.colorbar()
    plt.show()

# Paths to your real and fake images
real_image_path = '/content/real.jpg'
fake_image_path = '/content/fake.jpg'

# Perform DCT and shift high-frequency components to the center for both images
real_dct_img_high_freq_centered = perform_dct_high_freq_center(real_image_path)
fake_dct_img_high_freq_centered = perform_dct_high_freq_center(fake_image_path)

# Normalize the DCT images globally
normalized_images = global_normalize([real_dct_img_high_freq_centered, fake_dct_img_high_freq_centered])

# Enhance contrast of the normalized images
enhanced_images = enhance_contrast(normalized_images)

# Visualize the high-frequency centered, normalized, and contrast-enhanced DCT images
visualize_image(enhanced_images[0], "High-Frequency Centered DCT Real Image")
visualize_image(enhanced_images[1], "High-Frequency Centered DCT Fake Image")
```

Log Transformation: Applies log1p to the DCT coefficients to enhance the visibility of low-magnitude coefficients.
Global Normalization: Normalizes the log-transformed images.
Adaptive Histogram Equalization: Uses CLAHE (cv2.createCLAHE) to enhance the contrast of the normalized images.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform DCT and shift zero-frequency component to the center
def perform_dct(image_path):
    # Load the image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"Image at {image_path} could not be loaded.")

    # Resize the image to 224x224 (size expected by ViT model)
    img = cv2.resize(img, (224, 224))

    # Normalize pixel values to range [0, 1]
    img_normalized = img.astype(np.float32) / 255.0

    # Perform DCT
    dct_img = cv2.dct(img_normalized)

    # Shift the zero-frequency component to the center
    dct_img_centered = np.fft.fftshift(dct_img)

    return dct_img_centered

# Function to apply log transformation to the DCT coefficients
def log_transform(dct_images):
    log_transformed_images = []
    for image in dct_images:
        # Apply log transformation
        log_img = np.log1p(np.abs(image))
        log_transformed_images.append(log_img)
    return log_transformed_images

# Function to normalize images globally
def global_normalize(images):
    all_min = min(np.min(image) for image in images)
    all_max = max(np.max(image) for image in images)

    normalized_images = []
    for image in images:
        normalized_img = (image - all_min) / (all_max - all_min) * 255
        normalized_img = np.clip(normalized_img, 0, 255).astype(np.uint8)
        normalized_images.append(normalized_img)

    return normalized_images

# Function to enhance contrast using adaptive histogram equalization
def enhance_contrast(images):
    enhanced_images = []
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    for image in images:
        enhanced_img = clahe.apply(image)
        enhanced_images.append(enhanced_img)
    return enhanced_images

# Function to visualize an image
def visualize_image(image, title):
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.colorbar()
    plt.show()

# Paths to your real and fake images
real_image_path = '/content/real.jpg'
fake_image_path = '/content/fake.jpg'

# Perform DCT with center shift on both images
real_dct_img_centered = perform_dct(real_image_path)
fake_dct_img_centered = perform_dct(fake_image_path)

# Apply log transformation to the DCT images
log_transformed_images = log_transform([real_dct_img_centered, fake_dct_img_centered])

# Normalize the DCT images globally
normalized_images = global_normalize(log_transformed_images)

# Enhance contrast of the normalized images
enhanced_images = enhance_contrast(normalized_images)

# Visualize the centered, log-transformed, normalized, and contrast-enhanced DCT images
visualize_image(enhanced_images[0], "Centered Log-Transformed DCT Real Image")
visualize_image(enhanced_images[1], "Centered Log-Transformed DCT Fake Image")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform DCT and shift zero-frequency component to the center
def perform_dct(image_path):
    # Load the image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"Image at {image_path} could not be loaded.")

    # Resize the image to 224x224 (size expected by ViT model)
    img = cv2.resize(img, (224, 224))

    # Normalize pixel values to range [0, 1]
    img_normalized = img.astype(np.float32) / 255.0

    # Perform DCT
    dct_img = cv2.dct(img_normalized)

    # Shift the zero-frequency component to the center
    dct_img_centered = np.fft.fftshift(dct_img)

    return dct_img_centered

# Function to apply log transformation to the DCT coefficients
def log_transform(dct_images):
    log_transformed_images = []
    for image in dct_images:
        # Apply log transformation
        log_img = np.log1p(np.abs(image))
        log_transformed_images.append(log_img)
    return log_transformed_images

# Function to normalize images globally
def global_normalize(images):
    all_min = min(np.min(image) for image in images)
    all_max = max(np.max(image) for image in images)

    normalized_images = []
    for image in images:
        normalized_img = (image - all_min) / (all_max - all_min) * 255
        normalized_img = np.clip(normalized_img, 0, 255).astype(np.uint8)
        normalized_images.append(normalized_img)

    return normalized_images

# Function to apply contrast stretching
def contrast_stretch(images):
    stretched_images = []
    for image in images:
        min_val = np.min(image)
        max_val = np.max(image)
        stretched_img = (image - min_val) / (max_val - min_val) * 255
        stretched_img = np.clip(stretched_img, 0, 255).astype(np.uint8)
        stretched_images.append(stretched_img)
    return stretched_images

# Function to enhance contrast using adaptive histogram equalization
def enhance_contrast(images):
    enhanced_images = []
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    for image in images:
        enhanced_img = clahe.apply(image)
        enhanced_images.append(enhanced_img)
    return enhanced_images

# Function to visualize an image with colormap
def visualize_image(image, title):
    plt.imshow(image, cmap='viridis')  # Change colormap to 'viridis'
    plt.title(title)
    plt.axis('off')
    plt.colorbar()
    plt.show()

# Paths to your real and fake images
real_image_path = '/content/real.jpg'
fake_image_path = '/content/fake.jpg'

# Perform DCT with center shift on both images
real_dct_img_centered = perform_dct(real_image_path)
fake_dct_img_centered = perform_dct(fake_image_path)

# Apply log transformation to the DCT images
log_transformed_images = log_transform([real_dct_img_centered, fake_dct_img_centered])

# Normalize the DCT images globally
normalized_images = global_normalize(log_transformed_images)

# Apply contrast stretching to the normalized images
stretched_images = contrast_stretch(normalized_images)

# Enhance contrast of the stretched images
enhanced_images = enhance_contrast(stretched_images)

# Visualize the centered, log-transformed, normalized, stretched, and contrast-enhanced DCT images
visualize_image(enhanced_images[0], "Enhanced DCT Transformed Real Image")
visualize_image(enhanced_images[1], "Enhanced DCT Transformed Fake Image")
```

# USING FFT  Centers the zero-frequency component of the FFT for visualization.




```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to center the zero-frequency component using FFT
def center_zero_frequency(image):
    # Compute the FFT of the image
    fft_img = np.fft.fft2(image)

    # Shift the zero-frequency component to the center
    fft_shifted_img = np.fft.fftshift(fft_img)

    # Compute the magnitude spectrum for visualization
    magnitude_spectrum = 20 * np.log(np.abs(fft_shifted_img))

    return fft_shifted_img, magnitude_spectrum

# Function to visualize the original and centered FFT images
def visualize_fft(original_img, centered_img, magnitude_spectrum, title):
    plt.figure(figsize=(10, 5))

    # Original image
    plt.subplot(1, 3, 1)
    plt.imshow(original_img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # Centered FFT image
    plt.subplot(1, 3, 2)
    plt.imshow(np.abs(centered_img), cmap='gray')
    plt.title('Centered FFT')
    plt.axis('off')

    # Magnitude spectrum
    plt.subplot(1, 3, 3)
    plt.imshow(magnitude_spectrum, cmap='jet')
    plt.title('Magnitude Spectrum')
    plt.axis('off')

    plt.suptitle(title)
    plt.show()

# Paths to your real and fake images
real_image_path = '/content/real.jpg'
fake_image_path = '/content/fake.jpg'

# Read the real and fake images in grayscale
real_image = cv2.imread(real_image_path, cv2.IMREAD_GRAYSCALE)
fake_image = cv2.imread(fake_image_path, cv2.IMREAD_GRAYSCALE)

# Center the zero-frequency component using FFT for real image
centered_fft_real, magnitude_spectrum_real = center_zero_frequency(real_image)

# Center the zero-frequency component using FFT for fake image
centered_fft_fake, magnitude_spectrum_fake = center_zero_frequency(fake_image)

# Visualize the original and centered FFT images for real image
visualize_fft(real_image, centered_fft_real, magnitude_spectrum_real, 'Real Image')

# Visualize the original and centered FFT images for fake image
visualize_fft(fake_image, centered_fft_fake, magnitude_spectrum_fake, 'Fake Image')
```

# USING FFT
Extracts high-frequency components using a high-pass filter and shifts the high-frequency components back to the corners for visualization.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to extract high-frequency components using FFT
def extract_high_frequency(image, radius=30):
    # Compute the FFT of the image
    fft_img = np.fft.fft2(image)

    # Shift the zero-frequency component to the center
    fft_shifted_img = np.fft.fftshift(fft_img)

    # Get the dimensions of the image
    rows, cols = image.shape
    crow, ccol = rows // 2 , cols // 2     # center

    # Create a mask with a high-pass filter
    mask = np.ones((rows, cols), np.uint8)
    mask[crow-radius:crow+radius, ccol-radius:ccol+radius] = 0

    # Apply the mask to the shifted FFT
    fft_shifted_img_high_freq = fft_shifted_img * mask

    # Inverse FFT to get the high-frequency image
    ifft_shifted_img = np.fft.ifftshift(fft_shifted_img_high_freq)
    high_freq_img = np.fft.ifft2(ifft_shifted_img)
    high_freq_img = np.abs(high_freq_img)

    # Compute the magnitude spectrum for visualization
    magnitude_spectrum = 20 * np.log(np.abs(fft_shifted_img_high_freq) + 1)

    return high_freq_img, magnitude_spectrum

# Function to visualize the original and high-frequency images
def visualize_images(original_img, high_freq_img, magnitude_spectrum, title):
    plt.figure(figsize=(10, 5))

    # Original image
    plt.subplot(1, 3, 1)
    plt.imshow(original_img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # High-frequency image
    plt.subplot(1, 3, 2)
    plt.imshow(high_freq_img, cmap='gray')
    plt.title('High-Frequency Image')
    plt.axis('off')

    # Magnitude spectrum
    plt.subplot(1, 3, 3)
    plt.imshow(magnitude_spectrum, cmap='jet')
    plt.title('Magnitude Spectrum')
    plt.axis('off')

    plt.suptitle(title)
    plt.show()

# Paths to your real and fake images
real_image_path = '/content/real.jpg'
fake_image_path = '/content/fake.jpg'

# Read the real and fake images in grayscale
real_image = cv2.imread(real_image_path, cv2.IMREAD_GRAYSCALE)
fake_image = cv2.imread(fake_image_path, cv2.IMREAD_GRAYSCALE)

# Extract high-frequency components for real image
high_freq_real, magnitude_spectrum_real = extract_high_frequency(real_image)

# Extract high-frequency components for fake image
high_freq_fake, magnitude_spectrum_fake = extract_high_frequency(fake_image)

# Visualize the original and high-frequency images for real image
visualize_images(real_image, high_freq_real, magnitude_spectrum_real, 'Real Image')

# Visualize the original and high-frequency images for fake image
visualize_images(fake_image, high_freq_fake, magnitude_spectrum_fake, 'Fake Image')
```

# MFCC

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

# Function to compute MFCCs from spectrogram
def compute_mfcc(spectrogram, n_mfcc=13):
    mfccs = librosa.feature.mfcc(S=spectrogram, n_mfcc=n_mfcc)
    return mfccs

def visualize_mfcc(spectrogram, mfccs, title):
    plt.figure(figsize=(10, 6))

    plt.subplot(2, 1, 1)
    # Squeeze the spectrogram to remove the extra dimension
    librosa.display.specshow(librosa.amplitude_to_db(spectrogram.squeeze(), ref=np.max),
                             sr=22050, hop_length=512, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spectrogram')
    plt.tight_layout()

    plt.subplot(2, 1, 2)
    # Squeeze the mfccs array to remove the extra dimension
    librosa.display.specshow(mfccs.squeeze(), sr=22050, x_axis='time')
    plt.colorbar()
    plt.title('MFCCs')
    plt.tight_layout()

    plt.suptitle(title)
    plt.show()

# Paths to your real and fake images
real_image_path = '/content/real.jpg'
fake_image_path = '/content/fake.jpg'

# Read the real and fake images in grayscale
real_image = cv2.imread(real_image_path, cv2.IMREAD_GRAYSCALE)
fake_image = cv2.imread(fake_image_path, cv2.IMREAD_GRAYSCALE)

# Compute the spectrograms for real and fake images
spectrogram_real = librosa.feature.melspectrogram(y=real_image.astype(float), sr=22050, n_fft=2048, hop_length=512)
spectrogram_fake = librosa.feature.melspectrogram(y=fake_image.astype(float), sr=22050, n_fft=2048, hop_length=512)

# Compute MFCCs from the spectrograms
mfccs_real = compute_mfcc(spectrogram_real)
mfccs_fake = compute_mfcc(spectrogram_fake)

# Visualize the spectrograms and MFCCs for real image
visualize_mfcc(spectrogram_real, mfccs_real, 'Real Image')

# Visualize the spectrograms and MFCCs for fake image
visualize_mfcc(spectrogram_fake, mfccs_fake, 'Fake Image')
```
